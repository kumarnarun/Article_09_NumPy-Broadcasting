# Color to Greyscale conversion

```python
def rgb2gray(img):
    r = img[:, :, 0]
    g = img[:, :, 1]
    b = img[:, :, 2]
    return(0.2989 * r
          +0.5870 * g
          +0.1140 * b)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image2)

converted_image = rgb2gray(image2)

print(converted_image.shape)
plt.subplot(1, 2, 2)
plt.imshow(converted_image,
          cmap='gray')
plt.show()
```

# Color Transformation Example

```python
def do_normalise(img):
    return -np.log(1/((1 + img)/257) - 1)

def undo_normalise(img):
    return (1 + 1/(np.exp(-img) + 1) * 257).astype("uint8")
```

```python
def rotation_matrix(theta):
    return np.c_[[1, 0, 0],
                [0, np.cos(theta), -np.sin(theta)],
                [0, np.sin(theta), np.cos(theta)]]
```

```python
img1 = do_normalise(image1)
plt.imshow(img1)
plt.show()
```

```python
img2 = np.einsum("ijk,lk->ijl", image1, rotation_matrix(90))
plt.imshow(img2)
plt.show()
```

```python
img3 = do_normalise(image1)
plt.imshow(img3)
plt.show()
```

```python
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots(figsize=(5, 8))
```

```python
def update(i):
    imge1 = do_normalise(image1)
    imge2 = np.einsum("ijk,lk->ijl", image2, rotation_matrix(i * np.pi))
    imge3 = undo_normalise(image1)
    
    ax.imshow(imge3)
    ax.set_title("Angle: {0}*pi/10".format(i), fontsize=45)
```

```python
anim = FuncAnimation(fig, update,frames=np.arange(0, 40), interval=10)
anim.save('animated.gif',dpi=80, writer='pillow')
plt.close()
```

```python
plt.imshow(image2, cmap='gray')
plt.show()
```

```python
def threshold(img, thresh=127):
    return((img > thresh) * 255).astype("uint8")

def reverse_threshold(img, thresh=127):
    return((img < thresh) * 255).astype("uint8")
```

# Gradient

```python
def gradient(img, reverse=False):
    cols=img.shape[1]
    if reverse:
        C = np.linspace(1, 0, cols)
    else:
        C = np.linspace(0, 1, cols)
    C = np.dstack((C, C, C))
    print(C.shape)
    return(C * img).astype("uint8")
```

```python
plt.imshow(gradient(image2))
plt.show()
```

```python
plt.imshow(gradient(image2, True))
plt.show()
```

# Max RGB Filter

```python
def max_rgb(img):
    r = img[:, :, 0]
    g = img[:, :, 1]
    b = img[:, :, 2]
    
    M = np.maximum(np.maximum(r, g), b)
    
    r[r < M] = 0
    g[g < M] = 0
    b[b < M] = 0
    
    return(np.dstack((r, g, b)))
```

# intensity Normalization

```python
def normalize(img):
    lmin = float(img.min())
    lmax = float(img.max())
    return np.floor((img-lmin)/(lmax-lmin)*255.0)
```

```python
plt.imshow(normalize(image2), cmap='gray')
plt.show()
```
