```python
import cv2
import numpy as np
import random

# Set parameters
n = 10  # number of images to generate
image_size = 256  # size of the image in pixels
circle_radius = image_size//2  # radius of the circle
pinhole_radius = 3  # radius of the pinholes
circle_color = (0, 0, 255)  # red circle color  
pinhole_color = (0, 0, 0)  # color of the pinholes

# Generate images
for i in range(n):
    # Create the image
    image = np.zeros((image_size, image_size, 3), dtype=np.uint8)
    image[:] = circle_color

    # Generate the positions of the pinholes
    positions = [(int(circle_radius*np.cos(theta))+circle_radius, int(circle_radius*np.sin(theta))+circle_radius) 
                 for theta in sorted(random.sample(range(360), 3))]

    # Draw the pinholes on the image
    for position in positions:
        cv2.circle(image, position, pinhole_radius, pinhole_color, -1)

    # Draw the circle on top of the pinholes
    center = (image_size//2, image_size//2)
    cv2.circle(image, center, circle_radius-pinhole_radius, circle_color, thickness=2)

    # Save the image
    cv2.imwrite(f"image_{i}.png", image)
```
